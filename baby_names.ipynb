```python
years = range(1880, 2013)
pieces = []
columns = ['name', 'sex', 'births']

for year in years:
    path = 'yob%d.txt' % year
    frame = pd.read_csv(path, names = columns)
    
    frame['year'] = year
    pieces.append(frame)

#concatenate everything into a single DataFrame
names = pd.concat(pieces, ignore_index = True)  
```

```python
names
```

### Now, let's look at birth rates per year

```python
total_births = names.pivot_table('births', rows ='year', cols ='sex', aggfunc=sum)
```

```python
total_births.tail()
```

```python
plt.xkcd()
total_births.plot(title='Total Births since 1880', figsize=(12,5), colormap = 'cool_r')
```

```python
def add_prop(group):
    # Interget diviion floors
    births = group.births.astype(float)
    
    group['prop'] = births / births.sum()
    return group
names = names.groupby(['year','sex']).apply(add_prop)
```

### Now, let's extract the top 1000 names for each sex/year combination

```python
def get_top1000(group):
    return group.sort_index(by='births', ascending=False)[:1000]
grouped = names.groupby(['year','sex'])
top1000 = grouped.apply(get_top1000)
```

```python
def get_quantile_count(group, q=0.5):
    group = group.sort_index(by='prop', ascending = False)
    return group.prop.cumsum().searchsorted(q) + 1
```

```python
diversity = top1000.groupby(['year', 'sex']).apply(get_quantile_count)
```

```python
diversity = diversity.unstack('sex')
```

```python
diversity.head()
```

### Boy names that became girl names (and vice versa)

```python
all_names = top1000.name.unique()
```

```python
mask = np.array(['lesl' in x.lower() for x in all_names])
```

```python
lesley_like = all_names[mask]
```

```python
lesley_like
```

```python
filtered = top1000[top1000.name.isin(lesley_like)]
```

```python
filtered.groupby('name').births.sum()
```

```python
table = filtered.pivot_table('births', rows ='year', cols ='sex', aggfunc='sum')
```

```python
table = table.div(table.sum(1), axis =0)
```

```python
table.tail()
```

```python
table.plot(figsize(12,6), title = 'Relative freq of name uniqness', colormap='cool_r') 
```
